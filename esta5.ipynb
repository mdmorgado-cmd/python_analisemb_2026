```python
import pandas as pd
dados = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'
df = pd.read_csv(dados, encoding='latin-1', sep=';')

df_aaapai_cisp = df.groupby('cisp') ['aaapai'].sum().reset_index()
df_aaapai_cisp = df_aaapai_cisp.sort_values(by='aaapai', ascending=False)
df_aaapai_cisp 

```
