```python
import pandas as pd
import numpy as np
url = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'
df = pd.read_csv(url, encoding='latin-1', sep=';')
df_capital = df.loc [df['munic'] == 'Rio de Janeiro']

df_capital

df_cisp_capital = df.groupby('cisp') ['roubo_veiculo'].sum().reset_index()

df_cisp_capital = df_cisp_capital.sort_values(by='roubo_veiculo',  ascending=False)

df_cisp_capital

df_cisp_capital = df_capital.groupby('cisp') ['roubo_veiculo'].sum().reset_index()

df_cisp_capital = df_cisp_capital.sort_values(by='roubo_veiculo',  ascending=False)

array_roubo_veic = np.array(df_cisp_capital['roubo_veiculo'])
df_cisp_capital
print(array_roubo_veic)

```

```python

media = np.mean(array_roubo_veic)
mediana = np.median(array_roubo_veic)

print(f'O valor mediana é {mediana}')
print(f'O valor media é {media}')
```

```python
q1 = np.percentile(array_roubo_veic, 25)
q2 = np.percentile(array_roubo_veic, 50)
q3 = np.percentile(array_roubo_veic, 75)
print(f'Q1: {q1}')
print(f'Q2: {q2}')
print(f'Q3: {q3}')

df_mais_roubos = df_cisp_capital.loc[df_cisp_capital['roubo_veiculo'] > q3]

print(f'delegacias com menos roubos de veículo: ')

df_mais_roubos
```

```python
df_menos_roubos = df_cisp_capital.loc[df_cisp_capital['roubo_veiculo'] < q1]


print(f'delegacias com menos roubos de veículo:')

print (df_menos_roubos)
```
