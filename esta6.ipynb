```python
import pandas as pd
import numpy as np
url = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'
df = pd.read_csv(url, encoding='latin-1', sep=';')
df
```

```python
df_capital = df.loc [df['munic'] == 'Rio de Janeiro']

df_capital
```

```python
df_cisp_capital = df.groupby('cisp') ['furto_celular'].sum().reset_index()

df_cisp_capital = df_cisp_capital.sort_values(by='furto_celular',  ascending=False)

df_aaapai_cisp = df.groupby('cisp') ['aaapai'].sum().reset_index()

df_cisp_capital
```

```python
df_cisp_capital = df_capital.groupby('cisp') ['furto_celular'].sum().reset_index()

df_cisp_capital = df_cisp_capital.sort_values(by='furto_celular',  ascending=False)

array_furto = np.array(df_cisp_capital['furto_celular'])
df_cisp_capital
print(array_furto)
```

```python
maximo = np.max(array_furto)
minimo = np.min(array_furto)
media = np.mean(array_furto)
mediana = np.median(array_furto)

print(f'O valor maximo é {maximo}')
print(f'O valor minimo é {minimo}')
print(f'O valor mediana é {mediana}')
print(f'O valor media é {media}')
q1 = np.percentile(array_furto, 25)
q2 = np.percentile(array_furto, 50)
q3 = np.percentile(array_furto, 75)
print(f'Q1: {q1}')
print(f'Q2: {q2}')
print(f'Q3: {q3}')
```

```python
df_top25 = df_cisp_capital.loc[df_cisp_capital['furto_celular'] > q3]

print(df_top25)
```
